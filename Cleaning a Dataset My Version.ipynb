```python
# Scenario: given that the column MIN_GROUND_TEMP is stated that it is
# measured only every 6 hours, how can I check this?
```

```python
# Start by making a list of all positions in the column where I suspect that 
# there should be a number.
# Use range(5, len(df), 6) to generate a list of numbers starting with 5,
# up to the length of the DataFrame, with a step size of 6:
range(5, len(df), 6)
```

```python
# This can be made into a pandas series to display the contents (not actually
# necessary to do):
pd.Series(range(5, 8784, 6))
```

```python
#Store the list in a variable:
every_6th_row = pd.Series(range(5, 8784, 6))
```

```python
# This series can be used to retrieve rows from the data, so take the
# MIN_TEMP_GROUND column, and index it with the Series of positions.
# This will select every sixth row from this column:
df['MIN_TEMP_GROUND'][every_6th_row]
```

```python
# Check for a value that is NOT NaN in each cell:
df['MIN_TEMP_GROUND'][every_6th_row].notnull()
```

```python
# Summarise that by calling the function all(). This will return True, if
# all cells contain a value that is NOT NaN:
df['MIN_TEMP_GROUND'][every_6th_row].notnull().all()
```

```python
# Scenario: check the opposite hypothesis - are all other rows empty?
```

```python
# Drop every 6th row by using the drop() function:
df['MIN_TEMP_GROUND'].drop(every_6th_row)
```

```python
# Check that all remaining values are null or not:
df['MIN_TEMP_GROUND'].drop(every_6th_row).isnull()
```

```python
# Check that all values are True using all():
df['MIN_TEMP_GROUND'].drop(every_6th_row).isnull().all()
```

<h2>Function fillna()</h2>

```python
# function fillna() can be used to fill NaN values:
df['MIN_TEMP_GROUND'].fillna(0)
```

```python
# parameter 'method' can be set to 'ffil', which takes each measurement, and
# fills the NaN values after that with the last non-NaN value, until a new
# NaN value is found, and continue with that value:
df['MIN_TEMP_GROUND'].fillna(method='ffill')
```

```python
# Can use bfill to fill backwards:
df['MIN_TEMP_GROUND'].fillna(method='bfill')
```
